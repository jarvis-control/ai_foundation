```python
import pandas as pd
```

```python
mydataset = {
  'cars': ["Aston Martin", "BMW", "Volvo", "Ford", "Tesla", "Ferrari", "Bugatti"],
  'counts': [3, 7, 4, 2, 4, 5, 8]
}

df = pd.DataFrame(mydataset)
print(df)
```

```python
# columns - shows column names
df.columns
```

```python
# dtypes - shows the data type of each column
df.dtypes
```

```python
# describe() - gives summary statistics. By default, it focuses on numerical columns
# describe(include="all") - includes text columns too
df.describe()
```

```python
# iloc - selects by integer position. Think: position number
df.iloc[2:]
```

```python

# loc - selects by label. Think: row label and column label
df.loc[2:5, "cars"]
```

```python
data_with_missing = {
    "cars": ["BMW", "Volvo", None, "Tesla", "Ferrari", None, "Ford"],
    "counts": [3, 7, None, 8, 1, 9, None]
}

df_missing = pd.DataFrame(data_with_missing)

print(df_missing)
```

```python
# isnull() -
df_missing.isnull()
```

```python
# sum() -
df_missing.isnull().sum()
```

```python
# dropna() - removes missing data. The behavior depends on what you ask it to remove
# dropna(axis=0) - means columns. Since both cars and counts contain missing values, both columns may disappear
# dropna(axis=1) -

df_missing.dropna(axis=0)
```

```python
# dropna(subset=['counts']) - this removes only rows where counts is missing. If cars is missing but counts exists, that row stays
df_missing.dropna(subset=['counts'])
```

```python
# fillna() - replaces missing values
# This replaces missing car names with "unknown" and missing counts with 0
df_missing.fillna({"cars": "unkown", "counts": 0})
```
