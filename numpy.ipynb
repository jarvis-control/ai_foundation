```python
import numpy as np

# shape - Shows the structure of an array (rows, columns, etc.)
arr = np.array([1, 3, 5, 7])
print('numpy_array:', arr)

arr.shape
```

```python
# shape - Shows the structure of an array (rows, columns, etc.)
arr2 = np.array([[1, 3, 5, 7], [4, 6, 8, 9]])
print('numpy_arr2:', arr2)

arr2.shape
```

```python
# dtype - Shows the data type stored in the array
a = np.array([1, 2, 3])
print(a.dtype)
```

```python
# reshape() - Changes the layout/shape of an array without changing its values
a = np.array([1, 2, 3, 4, 5, 6])
b = a.reshape(2, 3)
print('shaped_array:', b)
```

```python
# astype() - Converts array values to another data type
a = np.array([1, 2, 3])
b = a.astype(float)
print('float:', b)
```

```python
# np.zeros() - Creates an array filled with zeros
# np.ones() - Creates an array filled with ones
# => dummy data
z = np.zeros((2, 3))
o = np.ones((2, 3))

print('zeros:', z)
print('ones:', o)
```

```python
# ndim - Shows how many dimensions an array has
a = np.array([[1, 2], [3, 4]])
print('ndim:', a.ndim)
```

```python
# arange() - Creates a sequence of evenly spaced numbers
a = np.arange(1, 7)
print('arange:', a)
```

```python
# sum() - Adds all values in an array
# mean() - Calculates the average of all values
# max() - Finds the largest value in an array
# min() - Finds the smallest value in an array
a = np.array([1, 2, 3, 4])

print('sum:', a.sum())
print('mean:', a.mean())
print('max:', a.max())
print('min:', a.min())
```

```python
# flatten() - Converts a multi-dimensional array into a single row of values
a = np.array([[1, 2], [3, 4]])
print('flatten:', a.flatten())
```

```python
# T (transpose) - Swaps rows and columns of an array
a = np.array([[1, 2], [3, 4]])
print('transpose:', a.T)
```

```python
# dot() - Performs dot product or matrix multiplication.
# @ - Another operator for matrix multiplication (same purpose as .dot())
a = np.array([[1, 2], [3, 4]])
b = np.array([[2, 0], [1, 2]])
print('dot/multiplication:', a.dot(b))
```

```python
# random.rand()
a = np.random.rand(2, 3) # => 0.0 and 1.0 / 0.0 included | 1.0 excluded
print('random_rand:', a)
```
